## Text Classification with NLTK

### Step 1:
### Create a list of tuples for all movie reviews. Each tuple contains:
### The list of words in the current review
### The category of review e.g. Positive / Negative

```python
import nltk
import random
from nltk.corpus import movie_reviews

list_tuples_for_all_movie_reviews = []

for category in movie_reviews.categories():
    for fileid in movie_reviews.fileids(category):
        list_tuples_for_all_movie_reviews.append((movie_reviews.words(fileid),category))

        
```

```python
list_tuples_for_all_movie_reviews
```

### Step 2:
### Shuffle the documents ( to mix Positive / Negative reviews)
### Create a list of all words in all documents

```python
#Shuffle the documents ( to mix Positive / Negative reviews)
random.shuffle(list_tuples_for_all_movie_reviews)

#Create a list of all words in all documents
all_words = []

for w in movie_reviews.words():
    all_words.append(w.lower())
    
all_words[:10]
```

### Step 3:
### Create a Frequency Distribution object of all words
### Use the most_common() function to return the top x most common words
### We can provide a word as a key, to get the number of occurrences of that word


```python
from nltk import FreqDist
all_words = nltk.FreqDist(all_words)
```

### Nltk.FreqDist(all_words) returns a dictionary of each unique word (key), 
### and the number of occurences (value):

```python
all_words
```

```python
all_words.most_common(10)
```

```python
all_words['great']
```

### Step 4:
### Obtain a list of the top 3000 most common words, stripping out the
### number of occurrences


```python
most_common_3000 = []
for w in list(all_words.most_common(3000)):
    most_common_3000.append(w[0])
```

```python
most_common_3000[:10]
```

###  Step 5:
###  Create a method which will:
###  Accept all the words in the current document
###  Get a set of each word in the current document (no duplicates)
###  Create an empty dictionary
###  Iterate through each common word, and populate the dictionary with key value pairs which are:
###  Key – the current common word
###  Value – A True/False value indicating if the common word is present in the document

```python
def find_feature(document):
    unique_words_document = set(document)
    features = {}
    for w in most_common_3000:
        features[w] = (w in unique_words_document)
    return features
```

### The output of this method will be a dictionary containing each
### common word, and whether the word is present in the current
### document, e.g.:

```python
print(find_feature(movie_reviews.words('pos\cv000_29590.txt')))
```

### Step 6:
### We have calculated the features for one file
### Now we calculate features for all files and record whether the review was
### negative or positive:
### Use a list comprehension:


```python
features_for_all_files = [(find_feature(reviews), category) for (reviews, category) in list_tuples_for_all_movie_reviews]
```

```python
features_for_all_files[:1]
```

# Pickle
##  After you train your classifier, you can save it

```python
import pickle

training_set = features_for_all_files[:1500]
test_set = features_for_all_files[1500:]

#Train our classifier using the training set
classifier = nltk.NaiveBayesClassifier.train(training_set)

#Save the classifier
save_classifier = open('naivebayes.pickle','wb')
pickle.dump(classifier, save_classifier)
save_classifier.close()
```

## Using our pre-trained classifier

```python
#Load the classifier
classifier_f = open('naivebayes.pickle','rb')
classifier = pickle.load(classifier_f)
classifier_f.close()

#Classify using out test set with our classifier, calculate the accuracy
accuracy = nltk.classify.accuracy(classifier, test_set)

print('Naïve Bayes Classifier Accuracy = ', accuracy)
```

# Scikit-Learn with NLTK
## An example of one Scikit-Learn classifier (a Multinominal Naïve Bayes Classifier)

```python
from nltk.classify.scikitlearn import SklearnClassifier
from sklearn.naive_bayes import MultinomialNB

MNB_classifier = SklearnClassifier(MultinomialNB())
MNB_classifier.train(training_set)
MNB_accuracy = nltk.classify.accuracy(MNB_classifier,test_set)
print('MultinomialNB Accuracy = ', MNB_accuracy)
```

## Using BernoulliNB classifiers

```python
from sklearn.naive_bayes import MultinomialNB, BernoulliNB

BNB_classifier = SklearnClassifier(BernoulliNB())
BNB_classifier.train(training_set)
BNB_accuracy = nltk.classify.accuracy(BNB_classifier,test_set)
print('BernoulliNB Accuracy = ', BNB_accuracy)
```

## Using LogisticRegression classifiers

```python
from sklearn.linear_model import LogisticRegression

logisticRegression_classifier = SklearnClassifier(LogisticRegression())
logisticRegression_classifier.train(training_set)
logisticRegression_accuracy = nltk.classify.accuracy(logisticRegression_classifier,test_set)
print('Logistic Regression accuracy = ', logisticRegression_accuracy)
```

```python
from sklearn.linear_model import SGDClassifier

SGD_classifier = SklearnClassifier(SGDClassifier())
SGD_classifier.train(training_set)
SGD_accuracy = nltk.classify.accuracy(SGD_classifier,test_set)

print('SGDClassifier accuracy = ', SGD_accuracy)
```

```python
from sklearn.svm import SVC

SVC_classifier = SklearnClassifier(SVC())
SVC_classifier.train(training_set)
SVC_accuracy = nltk.classify.accuracy(SVC_classifier,test_set)

print('SVC accuracy = ', SVC_accuracy)
```

```python
from sklearn.svm import LinearSVC

LinearSVC_classifier = SklearnClassifier(LinearSVC())
LinearSVC_classifier.train(training_set)
LinearSVC_accuracy = nltk.classify.accuracy(LinearSVC_classifier,test_set)

print('LinearSVC accuracy = ', LinearSVC_accuracy)
```

```python
from sklearn.svm import NuSVC

NuSVC_classifier = SklearnClassifier(NuSVC())
NuSVC_classifier.train(training_set)
NuSVC_accuracy = nltk.classify.accuracy(NuSVC_classifier,test_set)

print('NuSVC accuracy = ', NuSVC_accuracy)
```
